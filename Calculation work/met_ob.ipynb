```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = np.loadtxt('/Users/yurab/Desktop/Python for university/Calculation methods/Calculation work/result.csv', delimiter=';')
```

```python
Ny, Nx = data.shape
x = np.linspace(0, 1, Nx)
y = np.linspace(0, 1, Ny)
X, Y = np.meshgrid(x, y)
```

```python
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, data, levels=20, cmap='viridis')
plt.title('Контурний графік рішення')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(contour, label='u(x, y)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, data, cmap='viridis', edgecolor='none')
ax.set_title('3Д реалізація')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u(x, y)')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='u(x, y)')
plt.tight_layout()
plt.show()
```
