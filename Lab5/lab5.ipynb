```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("error_data.csv")

plt.figure(figsize=(10, 6))
plt.plot(data["x"], data["error"], label="Похибка |f(x) - P(x)|", color="crimson")

plt.title("Графік похибки інтерполяції Ньютона")
plt.xlabel("x")
plt.ylabel("Похибка")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
